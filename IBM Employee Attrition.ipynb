```python
#converting categorical columns to numerical coding
df['Gender'].replace(['Female','Male'],[1,0],inplace=True)
df['BusinessTravel'].replace(['Travel_Rarely','Travel_Frequently','Non-Travel'],[2,1,0],inplace=True)
df['Department'].replace(['Sales','Research & Development','Human Resources'],[0,1,2],inplace=True)
df['EducationField'].replace(['Life Sciences','Other','Medical','Marketing','Technical Degree',
 'Human Resources'],[1,6,2,3,4,5],inplace=True)
df['JobRole'].replace(['Sales Executive','Research Scientist','Laboratory Technician',
 'Manufacturing Director','Healthcare Representative','Manager',
 'Sales Representative','Research Director','Human Resources'],[1,2,3,4,5,6,7,8,9],inplace=True)
df['MaritalStatus'].replace(['Single','Married','Divorced'],[1,2,0],inplace=True)
df['OverTime'].replace(['Yes','No'],[1,0],inplace=True)
```

##### imoprting libraries for ML

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

##### split data into features and label

```python
X = df.drop('Attrition',axis = 1)
Y = df.Attrition
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=45)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
df.Attrition.value_counts()
```

we can observe that this data is inbalanced

```python
#for balncing our data
from imblearn.over_sampling import SMOTE
```

```python
smpl = SMOTE()
features,labels = smpl.fit_resample(X,Y)
sclr = preprocessing.MinMaxScaler()
sclr_data = pd.DataFrame(sclr.fit_transform(features),columns = features.columns)
```

```python
labels.value_counts()
```

```python
XB_train,XB_test,YB_train,YB_test = train_test_split(sclr_data,labels,test_size=0.2,random_state=45)
```

```python
XB_train.shape,XB_test.shape,YB_train.shape,YB_test.shape
```

##### using algorithms for both unbalanced and balanced data

#### Logistic Regression

```python
print("Logistic Regression")
log = LogisticRegression().fit(X_train,Y_train)
acc_log = log.score(X_test,Y_test)*100
print(acc_log)

print("----------------------")

log2 = LogisticRegression().fit(XB_train,YB_train)
acc_log2 = log2.score(XB_test,YB_test)*100
print(acc_log2)
```

#### Random Forest

```python
print("Random Forest")
ran = RandomForestClassifier().fit(X_train,Y_train)
acc_rn = ran.score(X_test,Y_test)*100
print(acc_rn)

print("----------------------")

ran2 = RandomForestClassifier().fit(XB_train,YB_train)
acc_rn2 = ran2.score(XB_test,YB_test)*100
print(acc_rn2)
```

#### Gradient Boosting

```python
print("Gradient boosting")
GB = GradientBoostingClassifier().fit(X_train,Y_train)
acc_gb = GB.score(X_test,Y_test)*100
print(acc_gb)

print("----------------------")

GB2 = GradientBoostingClassifier().fit(XB_train,YB_train)
acc_gb2 = GB2.score(XB_test,YB_test)*100
print(acc_gb2)
```

#### KNN

```python
print("KNN")
knn = KNeighborsClassifier().fit(X_train,Y_train)
acc_knn = knn.score(X_test,Y_test)*100
print(acc_knn)

print("----------------------")

knn2 = KNeighborsClassifier().fit(XB_train,YB_train)
acc_knn2 = knn2.score(XB_test,YB_test)*100
print(acc_knn2)
```

#### Decision Tree

```python
print("Decision Tree")
dt = DecisionTreeClassifier().fit(X_train,Y_train)
acc_dt = dt.score(X_test,Y_test)*100
print(acc_dt)

print("----------------------")

dt2 = DecisionTreeClassifier().fit(XB_train,YB_train)
acc_dt2 = dt2.score(XB_test,YB_test)*100
print(acc_dt2)
```

#### Navie Bayes

```python
print("Navie Bayes")
nb = GaussianNB().fit(X_train,Y_train)
acc_nb = nb.score(X_test,Y_test)*100
print(acc_nb)

print("----------------------")

nb2 = GaussianNB().fit(XB_train,YB_train)
acc_nb2 = nb2.score(XB_test,YB_test)*100
print(acc_nb2)
```

#### SVC

```python
print("SVC")
svc = SVC().fit(X_train,Y_train)
acc_svc = svc.score(X_test,Y_test)*100
print(acc_svc)

print("----------------------")

svc2 = SVC().fit(XB_train,YB_train)
acc_svc2 = svc2.score(XB_test,YB_test)*100
print(acc_svc2)
```

###### summerised all algorithms with that accuracy score of balanced and unbalanced data

```python
result = {
    "Algorithm":["LogisticRegression", "DecisionTreeClassifier", "KNeighborsClassifier", "GaussianNB", "SVC",
                 "RandomForestClassifier","GradientBoosting"],
    "Unbalanced_acc" : [acc_log,acc_dt,acc_knn,acc_nb,acc_svc,acc_rn,acc_gb],
    "Balanced_acc" : [acc_log2,acc_dt2,acc_knn2,acc_nb2,acc_svc2,acc_rn2,acc_gb2]}

print("Accuracy Score of all models Balanced data and unbalanced data")

Accuracy = pd.DataFrame(result)
Accuracy
```
